```python
import pandas as pd

data = {
    "Name" : ["Amit", "Rishi", "Nikhil", "Navya", "Santhosh"],
    "Maths": [78,81,92,91,95],
    "Science": [91,92,87,89,91],
    "English": [74,88,86,80,89]
}

df = pd.DataFrame(data)
df
```

```python
df["Total"] = df["Maths"] + df["Science"] + df["English"]
df["Average"] = df["Total"] / 3
df
```

```python
df.loc[df["Total"].idxmax()]
```

```python
df[df["Average"] > 85]
```

```python
df[["Maths", "Science", "English"]].describe()
```

```python
df.sort_values(by= "Total", ascending=False)
```

```python
df["Rank"] = df["Total"].rank(ascending=False)
df.sort_values("Rank")
```

```python
df["Result"] = df["Average"].apply(
    lambda x: "Pass" if x >= 85 else "Fail"
)
df
```

```python
print(df.loc[df["Maths"].idxmax(),["Name", "Maths"]])
print(df.loc[df["Science"].idxmax(),["Name", "Science"]])
print(df.loc[df["English"].idxmax(),["Name", "English"]])
```

```python
df.to_csv("student_report.csv", index=False)
```
